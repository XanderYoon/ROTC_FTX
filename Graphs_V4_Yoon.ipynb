```python
### Get rid of the non-essential columns
### Display the dataframe again
```

```python
df = df.drop(['Brevities', 'Unnamed: 13', 'Unnamed: 15', 'Positions', 'Raters'], axis=1)
df.head()
```

```python
### Filter cadets per school
df_GIT = df[df.School == 'GIT']
df_KSU = df[df.School == 'KSU']
df_EMORY = df[df.School == 'EMORY']
df_ASC = df[df.School == 'ASC']
```

```python
### Count the number of cadets in each school
n = df['School'].value_counts()
n
```

```python
# Display the Assessment occurrences
total_grades = df['Overall Assessment'].value_counts()
total_grades
```

```python
#Organizing Assessments by school
KSU_grades = df_KSU['Overall Assessment'].value_counts()
GIT_grades = df_GIT['Overall Assessment'].value_counts()
ASC_grades = df_ASC['Overall Assessment'].value_counts()
EMORY_grades = df_EMORY['Overall Assessment'].value_counts()

#Displaying grade distributions into pi charts
fig, GBS = plt.subplots(1, 5, figsize=(25, 5))  #GBS = Grade By School
fig.suptitle('Grade Distribution by School')
total_grades.plot.pie(ylabel='Total', autopct='%1.1f%%', ax=GBS[0])
KSU_grades.plot.pie(ylabel='KSU', autopct='%1.1f%%', ax=GBS[1])
GIT_grades.plot.pie(ylabel='GIT', autopct='%1.1f%%', ax=GBS[2])
ASC_grades.plot.pie(ylabel='ASC', autopct='%1.1f%%', ax=GBS[3])
EMORY_grades.plot.pie(ylabel='EMORY', autopct='%1.1f%%', ax=GBS[4])
```

```python
#Organizing Assessments based on MS Level
MS1_grades = df[df['MS Level'] == 1]['Overall Assessment'].value_counts()
MS2_grades = df[df['MS Level'] == 2]['Overall Assessment'].value_counts()
MS3_grades = df[df['MS Level'] == 3]['Overall Assessment'].value_counts()
MS4_grades = df[df['MS Level'] == 4]['Overall Assessment'].value_counts()

#Displaying grade distributions into pi chars
fig, GBL = plt.subplots(1, 4, figsize=(25, 5))  #GBL = Grade By Level
fig.suptitle('Grade Distribution by Level')
MS1_grades.plot.pie(ylabel='MS 1', autopct='%1.1f%%', ax=GBL[0])
MS2_grades.plot.pie(ylabel='MS 2', autopct='%1.1f%%', ax=GBL[1])
MS3_grades.plot.pie(ylabel='MS 3', autopct='%1.1f%%', ax=GBL[2])
MS4_grades.plot.pie(ylabel='MS 4', autopct='%1.1f%%', ax=GBL[3])
```

```python
#Position v Grade
pos_grades = df.groupby('Rated Pos.')['Overall Assessment'].value_counts()
pos_grades = pos_grades.unstack(level=1)
pos_grades
```

```python
#Grade v Position
grades_pos = df.groupby('Overall Assessment')['Rated Pos.'].value_counts()
grades_pos = grades_pos.unstack(level=1)
grades_pos
```

```python
fig, GBP = plt.subplots(1, 2, figsize=(25, 5))  #GBP = Grade By Position
fig.suptitle('Grade Distribution by Position')

pos_grades.plot.bar(xlabel='POSITION', ax=GBP[0]) #Position v Grade
grades_pos.plot.bar(xlabel='GRADE', ax=GBP[1]) #Grade v Position
```

```python
#Grades by Rater
rater_grades = df.groupby('Rater')['Overall Assessment'].value_counts()
rater_grades = rater_grades.unstack(level=1)
rater_grades.plot(kind='bar')
```

```python
#"Decoding" Brevity code
cols = ['Sustain 1', 'Sustain 2', 'Sustain 3', 'Improve 1', 'Improve 2', 'Improve 3']
df[cols] = df[cols].replace(["AV", "EM", "WE", "DP", "HU"], "Character")
df[cols] = df[cols].replace(["MB", "FT", "CF", "RS", 'GR'], "Presence")
df[cols] = df[cols].replace(["MA", "SJ", "IN", "IT", "EX"], "Intellect")
df[cols] = df[cols].replace(["LD", "BT", "EI", "LE", "CO"], "Lead")
df[cols] = df[cols].replace(["CP", "PS", "DO", "SP"], "Develop")
df #what is GR? Put under Presence for now
```

```python
###SUSTAINS
#Subplot
fig, SB = plt.subplots(1,4, figsize=(25,5))
fig.suptitle('Sustain Data')

#Dataframes by Position/Rater/School and Sustains
pos_sustain = df.groupby('Rated Pos.')['Sustain 1' and 'Sustain 2' and 'Sustain 3'].value_counts()
rater_sustain = df.groupby('Rater')['Sustain 1' and 'Sustain 2' and 'Sustain 3'].value_counts()
school_sustain = df.groupby('School')['Sustain 1' and 'Sustain 2' and 'Sustain 3'].value_counts()

#Displaying data
total_sustain = df['Sustain 1' and 'Sustain 2' and 'Sustain 3'].value_counts().plot(kind="bar", xlabel="TOTAL", ax=SB[0])
pos_sustain.unstack().plot(kind='bar', xlabel="POSITION", ax=SB[1])
rater_sustain.unstack().plot(kind='bar', xlabel="RATER", ax=SB[2])
school_sustain.unstack().plot(kind='bar', xlabel="SCHOOL", ax=SB[3])
```

```python
###IMPROVES
#Subplot
fig, SB = plt.subplots(1,4, figsize=(25,5))
fig.suptitle('Improve Data')

#Dataframes by Position/Rater/School and Sustains
pos_sustain = df.groupby('Rated Pos.')['Improve 1' and 'Improve 2' and 'Improve 3'].value_counts()
rater_sustain = df.groupby('Rater')['Improve 1' and 'Improve 2' and 'Improve 3'].value_counts()
school_sustain = df.groupby('School')['Improve 1' and 'Improve 2' and 'Improve 3'].value_counts()

#Displaying data
total_sustain = df['Improve 1' and 'Improve 2' and 'Improve 3'].value_counts().plot(kind="bar", xlabel="TOTAL", ax=SB[0])
pos_sustain.unstack().plot(kind='bar', xlabel="POSITION", ax=SB[1])
rater_sustain.unstack().plot(kind='bar', xlabel="RATER", ax=SB[2])
school_sustain.unstack().plot(kind='bar', xlabel="SCHOOL", ax=SB[3])
```
